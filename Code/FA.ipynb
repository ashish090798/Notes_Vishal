```python
df.drop(['Arrival Delay in Minutes','Class'], axis=1, inplace=True)
```

```python
x =df[df.columns[6:20]] 
```

```python
#Model building
fa = FactorAnalyzer()
fa.fit(x)
ev, v = fa.get_eigenvalues()
plt.plot(range(1,x.shape[1]+1),ev)
fa = FactorAnalyzer(3, rotation='varimax')
fa.fit(x)
loads = fa.loadings_
print(loads)
```
