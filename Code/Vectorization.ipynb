```python
filename=r'C:\Users\Vish\Documents\Data\Dataset\metamorphosis.txt'
file=open(filename,'rt')
text=file.read()

from nltk import sent_tokenize
sentences = sent_tokenize(text)

from nltk.tokenize import word_tokenize
tokens = [word_tokenize(text) for text in sentences[0:2]]

from nltk.corpus import stopwords
stop_words = stopwords.words('english')
final_list=[]
for i in tokens:
    words = ' '.join([word for word in i if word.isalpha() and not word in stop_words])
    final_list.append(words)
    
Document1=final_list[0]
Document2=final_list[1]
Document1 = Document1.lower()
Document2 = Document2.lower()
print(Document1) 
print(Document2)

#REPRESENTATION FOR DOCUMENTS
from sklearn.feature_extraction.text import CountVectorizer
Count_vect = CountVectorizer()
Count_vect.fit([Document1, Document2])

Bow_Document1 = Count_vect.transform([Document1])
print("Representation Of Document1: ")
print(pd.DataFrame(Bow_Document1.toarray(),columns=Count_vect.get_feature_names()))

Bow_Document2 = Count_vect.transform([Document2])
print("Representation Of Document2: ") 
print(pd.DataFrame(Bow_Document2.toarray(),columns=Count_vect.get_feature_names()))

#REPRESENTATION FOR CORPUS
corpus=[Document1+Document2]
Count_vect.fit(corpus)
Bow_Corpus = Count_vect.transform(corpus)
print('Representation of Corpus:')
print(pd.DataFrame(Bow_Corpus.toarray(),columns=Count_vect.get_feature_names()))
```

```python
filename=r'C:\Users\Vish\Documents\Data\Dataset\metamorphosis.txt'
file=open(filename,'rt')
text=file.read()

from nltk import sent_tokenize
sentences = sent_tokenize(text)

from nltk.tokenize import word_tokenize
tokens = [word_tokenize(text) for text in sentences[0:2]]

from nltk.corpus import stopwords
stop_words = stopwords.words('english')
final_list=[]
for i in tokens:
    words = ' '.join([word for word in i if word.isalpha() and not word in stop_words])
    final_list.append(words)
      
Document1=final_list[0]
Document2=final_list[1]
Document1 = Document1.lower()
Document2 = Document2.lower()
print('Documents :')
print(Document1) 
print(Document2)

#REPRESENTATION FOR DOCUMENTS
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf_vect = TfidfVectorizer()
Tfidf_vect.fit([Document1, Document2])

Bow_Document1 = Tfidf_vect.transform([Document1])
print("Representation Of Document1: ")
print(pd.DataFrame(Bow_Document1.toarray(),columns=Tfidf_vect.get_feature_names()))

Bow_Document2 = Tfidf_vect.transform([Document2])
print("Representation Of Document2: ")
print(pd.DataFrame(Bow_Document2.toarray(),columns=Tfidf_vect.get_feature_names()))

#REPRESENTATION FOR CORPUS
corpus=[Document1+Document2]
Tfidf_vect.fit(corpus)
bog_corpus = Tfidf_vect.transform(corpus)
print('Representation of Corpus:')
print(pd.DataFrame(bog_corpus.toarray(),columns=Tfidf_vect.get_feature_names()))
```

```python
filename=r'C:\Users\Vish\Documents\Data\Dataset\metamorphosis.txt'
file=open(filename,'rt')
text=file.read()

from nltk import sent_tokenize
sentences = sent_tokenize(text)

from nltk.tokenize import word_tokenize
tokens = [word_tokenize(text) for text in sentences[0:2]]

from nltk.corpus import stopwords
stop_words = stopwords.words('english')
final_list=[]
for i in tokens:
    words = [word for word in i if word.isalpha() and not word in stop_words]
    final_list.append(words)

print('TOKENS :')
print(final_list)

#REPRESENTATION FOR CORPUS
from gensim.models import Word2Vec                  
word2vec = Word2Vec(sentences=final_list, vector_size=100, window=30, min_count=1,sg=1,epochs=50)
#sentences:iterable of iterables,optional--tokenized corpus
#corpus_file:str,optional--path to a corpus file in LineSentence format
#vector_size:int,optional,default=100--dimensionality of word vectors
#window:int,optional--max distance between current and predicted words within a sentence
#min_count:int,optional--ignores all words with frequency less than this
#epochs:int,optional,default=100--number of iterations over the corpus
#alpha:float, optional,default=0.025--The initial learning rate
#min_alpha:float,optional,default==0.0001--Learning rate will linearly drop to min_alpha as training progresses.
#with each iteration, to min_alpha. The learning rate determines how easily the hidden layer adjusts to new input. 
# Higher values will make training still effective with less data but will likely incorporate more noise/overfit
#sg:{0,1},optional,default=0--1 for skip gram and 0 for cbow.
#hs:{0,1},optional,default=0--If you set hs=1, word2vec will use hierarchical softmax to train the model. Otherwise it will use negative sampling. 
#The rule of thumb here is that hierarchical softmax works better for infrequent wordsas it builds a tree over the whole vocabulary.
#Negative sampling works better for frequent words, because it tries to differentiate signal from noise.
#negative:int,optional,default=5--if >0, negative sampling will be used, the int for negative specifies how many noise words 
#should be drawn. For negative sampling, the signal vs noise differentiation is done by comparing words from the context with 
#random words from outside the context.  A default value of 5 tells word2vec to pick 5 random words outside the context
#ns_exponent:float, optional,default=0.75 -- The exponent used to shape the negative sampling distribution. 
#ns_exponent tells word2vec which words to prefer when sampling. 1.0 means sampled words will have a proportional frequency 
#to the context words. ns_exponent=0.0 means no frequency preference and it samples all words equally, while negative values 
#will prefer lower frequency words. The default value of 0.75, which works well
#cbow_mean:{0, 1}, optional-- If 0, use the sum of the context word vectors. If 1, use the mean, only applies when cbow is used.
#seed:int, optional-- Seed for the random number generator. 
#Initial vectors for each word are seeded with a hash of the concatenation of word + str(seed). 
#workers:int,optional--use these many worker threads to train the model
#max_vocab_size:int, optional,default=40M-- maximum size (number of tokens) of the vocabulary. 
#Limits the RAM during vocabulary building. Used to control pruning of less common words. 
#If there are more unique words than this then prune the infrequent ones.
#Every 10 million word types need about 1GB of RAM. Set to None for no limit. #batch_words:int,optional--how many words to send at once for each worker thread.
#max_final_vocab:int,optional-– Limits the vocab to a target vocab size by automatically picking a matching min_count. 
# If the specified min_count is more than the calculated min_count, the specified min_count will be used. Set to None if not required.
# sample:float, optional-- The threshold for configuring which higher-frequency words are randomly downsampled. Useful range is (0, 1e-5).
#hashfxn:functional,optional--hash function to use to randomly intialize weights for increased training reproducibility
#trim_rule:function,optional--vocabulary trimming rule,specifies whether certain words should remain in the vocabulary, be 
#trimmed away or handled using the default (discard if word count <min count)

vocabulary = list(word2vec.wv.index_to_key)
print('VOCABULARY OF CORPUS :')
print(vocabulary)

v1 = word2vec.wv['restrictions'] 
print(v1)

similar_words = {search_term: [item[0] for item in word2vec.wv.most_similar([search_term], topn=5)]
                  for search_term in ['restrictions','whatsoever']}
print(similar_words)
# similar_words = word2vec.wv.most_similar('restrictions')
# print(similar_words)

# word2vec.save('Word2VecModel')
# model = Word2Vec.load('Word2VecModel')

from sklearn.manifold import TSNE
words = sum([[k] + v for k, v in similar_words.items()], [])
wvs = word2vec.wv[words]
tsne = TSNE(n_components=2, random_state=0, n_iter=10000, perplexity=2)
np.set_printoptions(suppress=True)
T = tsne.fit_transform(wvs)
labels = words
plt.figure(figsize=(14, 8))
plt.scatter(T[:, 0], T[:, 1], c='orange', edgecolors='r')
for label, x, y in zip(labels, T[:, 0], T[:, 1]):
    plt.annotate(label, xy=(x+1, y+1), xytext=(0, 0), textcoords='offset points')
```

```python
#BIGRAM
from gensim.models.phrases import Phrases,ENGLISH_CONNECTOR_WORDS
 
# load the text8 dataset
filename=r'C:\Users\Vish\Documents\Data\Dataset\metamorphosis.txt'
file=open(filename,'rt')
text=file.read()

from nltk import sent_tokenize
sentences = sent_tokenize(text)

from nltk.tokenize import word_tokenize
tokens = [word_tokenize(text) for text in sentences]

from nltk.corpus import stopwords
stop_words = stopwords.words('english')
final_list=[]
for i in tokens:
    words = [word for word in i if word.isalpha() and not word in stop_words]
    final_list.append(words)
print('TOKENS :')
print(final_list[0])

bigram_model = Phrases(final_list, min_count = 1, threshold = 1,connector_words=ENGLISH_CONNECTOR_WORDS)
#sentences:iterable of iterables,optional--tokenized corpus
#min_count:int,optional--ignores all words with frequency less than this
#threshold:float, optional-- represents a score threshold for forming the phrases (higher means fewer phrases). 
#a phrase of words a followed by b is accepted if the score of the phrase is greater than threshold. 
#max_vocab_size:int, optional,default=40M-- maximum size (number of tokens) of the vocabulary. 
#Limits the RAM during vocabulary building. Used to control pruning of less common words. 
#If there are more unique words than this then prune the infrequent ones.
#Every 10 million word types need about 1GB of RAM. Set to None for no limit. 
#to keep memory under control. The default of 40M needs about 3.6GB of RAM.Increase/decrease depending on memory you have.
#scoring:{”default” - original_scorer(),”npmi” - npmi_scorer()},optional--specifies how potential phrases are scored. scoring can be set with either a string 
#that refers to a built-in scoring function or with a function with the expected parameter names. Two built-in scoring functions 
#are available by setting scoring to a string.
#connector_words:set of str, optional--set of words that may be included within a phrase, without affecting its scoring. 
#No phrase can start nor end with a connector word; a phrase may contain any number of connector words in the middle.
#If your texts are in English, set connector_words=phrases.ENGLISH_CONNECTOR_WORDS.This will cause phrases to include 
#common English articles, prepositions and conjuctions, such as bank_of_america or eye_of_the_beholder.
#For other languages or specific applications domains, use custom connector_words that make sense 
#there: connector_words=frozenset("der die das".split()) etc
#delimiter:str, optional--Glue character used to join collocation tokens
print('BIGRAMS:')
print(bigram_model[final_list[0]]) 

from gensim.models import Word2Vec                  
word2vec_bigram = Word2Vec(sentences=bigram_model[final_list], vector_size=100, window=30, min_count=1,sg=1,epochs=50)

vocabulary = list(word2vec_bigram.wv.index_to_key)
# print('VOCABULARY OF CORPUS :')
# print(vocabulary)

v1 = word2vec_bigram.wv['Kafka_Translated'] 
print(v1)

similar_words = {search_term: [item[0] for item in word2vec_bigram.wv.most_similar([search_term], topn=5)]
                  for search_term in ['Kafka_Translated','Metamorphosis']}
print(similar_words)
# similar_words = word2vec.wv.most_similar('Kafka_Translated')
# print(similar_words)

# word2vec.save('Word2VecModel')
# model = Word2Vec.load('Word2VecModel')

from sklearn.manifold import TSNE
words = sum([[k] + v for k, v in similar_words.items()], [])
wvs = word2vec_bigram.wv[words]
tsne = TSNE(n_components=2, random_state=0, n_iter=10000, perplexity=2)
np.set_printoptions(suppress=True)
T = tsne.fit_transform(wvs)
labels = words
plt.figure(figsize=(14, 8))
plt.scatter(T[:, 0], T[:, 1], c='orange', edgecolors='r')
for label, x, y in zip(labels, T[:, 0], T[:, 1]):
    plt.annotate(label, xy=(x+1, y+1), xytext=(0, 0), textcoords='offset points')
```
